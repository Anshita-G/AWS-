```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
plt.figure(figsize=(10,10))
for i in range(100):
  plt.subplot(10,10,i+1)
  plt.axis('off')
  plt.imshow(x_train[i])
```

```python
x_train = x_train / 255
```

```python
batch_size = 32
dataset = tf.data.Dataset.from_tensor_slices(x_train).shuffle(1000)
dataset = dataset.batch(batch_size, drop_remainder = True).prefetch(1)
```

```python
codings_size= 30
generator = keras.models.Sequential([
    keras.layers.Dense(128 , activation= 'selu', input_shape=[codings_size]),
    keras.layers.Dense(256, activation= 'selu'),
    keras.layers.Dense(28*28, activation= 'selu'),
    keras.layers.Reshape([28,28])
])
generator.summary()
```

```python
discreminator = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28,28]),                                     
    #keras.layers.Dense(128 , activation= 'selu', input_shape=[codings_size]),
    keras.layers.Dense(256, activation= 'selu'),
    keras.layers.Dense(128, activation= 'selu'),
    keras.layers.Dense(1, activation= 'sigmoid')
])
discreminator.summary()
```

```python
gan = keras.models.Sequential([generator, discreminator])
gan.summary()
```

```python
discreminator.compile(loss ='binary_crossentropy' , optimizer= 'rmsprop')
discreminator.trainable = False
gan.compile(loss ='binary_crossentropy' , optimizer= 'rmsprop')
```

```python
def train_gan(gan , dataset, batch_size, codings_size,n_epochs = 100):
  generator, discreminator = gan.layers
  for epoch in range(n_epochs):
    for x_batch in dataset:
      noise = tf.random.normal(shape = [batch_size, codings_size])
      generated_images = generator(noise)
      x_batch = tf.cast(x_batch, tf.float32)

      x_fake_and_real = tf.concat([generated_images, x_batch], axis = 0)
      y1 = tf.constant([[0.]] * batch_size +[[1.]] * batch_size)

      discreminator.trainable = True
      discreminator.train_on_batch(x_fake_and_real, y1)

      noise = tf.random.normal(shape = [batch_size, codings_size])
      y2 = tf.constant(([[1.]] * batch_size))

      discreminator.trainable = False
      gan.train_on_batch(noise, y2)
  
```

```python
train_gan(gan, dataset, batch_size, codings_size)
```
